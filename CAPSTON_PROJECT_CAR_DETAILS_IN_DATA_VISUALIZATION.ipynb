# DATA VISUALIZATION

```python
# First Plot: Making a Line Chart of Selling Price Growth Trends
data.groupby(['year'])['selling_price'].sum().plot()
plt.show()
```

```python
plt.figure(figsize=(15,5))
data.groupby(['year'])['selling_price'].sum().plot()
plt.title('Total Selling Price in Every Year', loc='center', pad=20, fontsize=20, color='blue')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

Total Selling Price in Every Year - Breakdown by Seller Type

```python
# Multi Line Chart

data.groupby(['year','seller_type'])['selling_price'].sum().unstack().plot()
plt.title('Total Selling Price in Every Year - Breakdown by Seller Type', loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Total Selling(in Millions)', fontsize = 15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.gcf().set_size_inches(10, 5)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart
plt.clf()
plt.figure(figsize=(7,7))
data_sum = data.groupby('seller_type')['selling_price'].sum().reset_index()
data_sum.groupby('seller_type')['selling_price'].sum().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Total Selling Price in All Year - Breakdown by Seller Type', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Seller Type', fontsize=15)
plt.ylabel('Total Selling Price (in Millions)', fontsize=15)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=0)
plt.show()
```

```python
# Pie Chart
data_sum = data.groupby('seller_type')['selling_price'].sum().reset_index()

plt.figure(figsize=(7,7))
plt.pie(data_sum['selling_price'], labels = data_sum['seller_type'], autopct='%1.2f%%')
plt.title('Total Selling Price in All Year - Breakdown by Seller Type', loc='center', pad=30, fontsize=15, color='blue')
plt.show()
```

We can see that the individual seller_type dominates car sales in this case with 62,94 %, followed by Dealers and Trustmark Dealers. so we can say individual dealer sales account for the majority of car sales in this case, with the highest sales in 2017

Total Selling Price in Every Year - Breakdown by Owner

```python
# Multi Line Chart

data.groupby(['year','owner'])['selling_price'].sum().unstack().plot()
plt.title('Total Selling Price in Every Year - Breakdown by Owner', loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Total Selling(in Millions)', fontsize = 15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.gcf().set_size_inches(10, 5)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart
plt.clf()
plt.figure(figsize=(7,7))
data_sum = data.groupby('owner')['selling_price'].sum().reset_index()
data_sum.groupby('owner')['selling_price'].sum().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Total Selling Price in All Year - Breakdown by Owner', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Owner', fontsize=15)
plt.ylabel('Total Selling Price (in Millions)', fontsize=15)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=0)
plt.show()
```

```python
# Pie Chart
data_sum = data.groupby('owner')['selling_price'].sum().reset_index()

plt.figure(figsize=(7,7))
plt.pie(data_sum['selling_price'], labels = data_sum['owner'], autopct='%1.2f%%')
plt.title('Total Selling Price in All Year - Breakdown by Owner', loc='center', pad=30, fontsize=15, color='blue')
plt.show()
```

From the graph, we can conclude that the majority of car sales come from first owners with 77,49%, with the highest sales in 2017

Another Visualization

```python
# Membuat Multi-Bar Chart
data16_20 = data[(data['year']>= 2016)]

data16_20.groupby(['year', 'seller_type'])['selling_price'].sum().unstack().plot(kind='bar')
plt.title('Total Selling Price in 2016-2020 - Breakdown by Seller Type (in Millions)', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Selling Price (in Millions)', fontsize=12)
plt.legend(bbox_to_anchor=(1,1), shadow=True, title='Seller Type')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Membuat Multi-Bar Chart
data16_20 = data[(data['year']>= 2016)]

data16_20.groupby(['year', 'seller_type'])['selling_price'].sum().unstack().plot(kind='bar', width=0.8, figsize = (15,10))
plt.title('Total Selling Price in 2016-2020 - Breakdown by Seller Type (in Millions)', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Selling Price (in Millions)', fontsize=12)
plt.legend(bbox_to_anchor=(1,1), shadow=True, title='Seller Type')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Membuat Stacked Chart

data16_20.groupby(['seller_type','year'])['selling_price'].sum().sort_values(ascending=False).unstack().plot(kind='bar', stacked=True)
plt.title('Total Selling Price Every Sellr Type - Breakdown by Year (in Millions)', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Seller Type', fontsize=12)
plt.ylabel('Total Selling Price (in Millions)', fontsize=12)
plt.legend(bbox_to_anchor=(1,1), shadow=True, ncol=1, title='City')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
